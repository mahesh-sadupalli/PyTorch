### Initial settings in general:
1) Seed
2) Device 
3) Number of epochs
4) Learning rate
5) Batch size
6) Embedded dimesions 
7) Hidden dimensions 
8) Number of classes 
9) Vocabulary size

```python

random_seed = 42
torch.manual_seed(random_seed)

 
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")


num_epochs = 15


learning_rate = 0.005


batch_size = 128

emd_dim = 128

hidden_dim = 256

num_classes = 2

vocab_size = 20000

print(f"Seed: {random_seed} \nDevice using:{device} \nNumber of Epochs: {num_epochs}")
print(f"Learning Rate: {learning_rate} \nBatch Size:{batch_size}\nNumber of Epochs: {num_epochs}")
print(f"Embedded dimension: {emd_dim} \nHidden Dimension:{hidden_dim}\nNumber of Classes: {num_classes} \nVocabulary Size: {vocab_size}")

```
