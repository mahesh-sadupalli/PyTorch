```python
# Load raw data first (no transforms)
raw_dataset = datasets.FashionMNIST('./data', train=True, download=True, transform=None)

# Check for issues
def check_data_quality(dataset, num_samples=1000):
    print("=" * 50)
    print("DATA QUALITY CHECK")
    print("=" * 50)
    
    for i in range(min(num_samples, len(dataset))):
        img, label = dataset[i]
        img_array = np.array(img)
        
        # Check for common issues
        if img_array.shape != (28, 28):
            print(f"Image {i}: Unexpected shape {img_array.shape}")
        if label < 0 or label > 9:
            print(f"Image {i}: Invalid label {label}")
        if img_array.min() < 0 or img_array.max() > 255:
            print(f"Image {i}: Pixel values out of range [{img_array.min()}, {img_array.max()}]")
    
    # Check data statistics
    sample_img, _ = dataset[0]
    sample_array = np.array(sample_img)
    print(f"\nData Statistics:")
    print(f"Image shape: {sample_array.shape}")
    print(f"Data type: {sample_array.dtype}")
    print(f"Pixel range: [{sample_array.min()}, {sample_array.max()}]")
    print(f"Mean pixel value: {sample_array.mean():.2f}")
    print(f"Std pixel value: {sample_array.std():.2f}")

check_data_quality(raw_dataset)
```

```python
# Calculate mean and std of the dataset
def calculate_dataset_stats(dataset):
    # Load all data (careful with large datasets!)
    dataloader = torch.utils.data.DataLoader(dataset, batch_size=len(dataset))
    data = next(iter(dataloader))[0]  # Get all images
    
    mean = data.mean()
    std = data.std()
    return mean, std

# For Fashion-MNIST (pre-calculated):
# Mean  0.5, Std  0.5 for normalized [0,1] data

# Standard normalization transform
transform_normalized = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.5,), std=(0.5,))  # For grayscale (1 channel)
])

# What this does:
# normalized = (pixel - mean) / std
# Maps [0, 1] to approximately [-1, 1]

# Compare the effects:
normalized_dataset = datasets.FashionMNIST('./data', train=True, download=True,
                                          transform=transform_normalized)

tensor_img, _ = tensor_dataset[0]
norm_img, _ = normalized_dataset[0]

print(f"Before normalization: min={tensor_img.min():.2f}, max={tensor_img.max():.2f}")
print(f"After normalization: min={norm_img.min():.2f}, max={norm_img.max():.2f}")

# Visualize the difference
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(np.array(raw_img), cmap='gray')
axes[0].set_title('Original (0-255)')
axes[0].axis('off')

axes[1].imshow(tensor_img.squeeze(), cmap='gray')
axes[1].set_title('ToTensor (0-1)')
axes[1].axis('off')

axes[2].imshow(norm_img.squeeze(), cmap='gray')
axes[2].set_title('Normalized (-1 to 1)')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

```python
#Building CNN
import torch.nn as nn
import torch.nn.functional as F
class BasicCNN(nn.Module):
    def __init__(self):
        super(BasicCNN, self).__init__()
        
        # Convolutional layers
        self.conv1 = nn.Conv2d(1, 6, 5)     # Input: 1 channel, Output: 6 channels, Kernel: 5x5
        self.conv2 = nn.Conv2d(6, 16, 5)    # Input: 6 channels, Output: 16 channels, Kernel: 5x5
        
        # Pooling layer (used after each conv layer)
        self.pool = nn.MaxPool2d(2, 2)      # 2x2 pooling, stride=2
        
        # Fully connected layers
        self.fc1 = nn.Linear(16 * 4 * 4, 120)  # Flattened size -> 120 neurons
        self.fc2 = nn.Linear(120, 84)           # 120 -> 84 neurons  
        self.fc3 = nn.Linear(84, 10)            # 84 -> 10 classes
        
    def forward(self, x):
        # First conv block
        x = self.pool(F.relu(self.conv1(x)))   # Conv -> ReLU -> Pool
        
        # Second conv block  
        x = self.pool(F.relu(self.conv2(x)))   # Conv -> ReLU -> Pool
        
        # Flatten for FC layers
        x = x.view(-1, 16 * 4 * 4)  # Reshape to (batch_size, features)
        
        # Fully connected layers
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)  # No activation here - CrossEntropyLoss handles it
        
        return x

# Create model and check architecture
model = BasicCNN()
print(model)

# Count parameters
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nTotal parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")
```

```python
def trace_dimensions(model, input_size=(1, 1, 28, 28)):
    """Trace how dimensions change through the network"""
    
    x = torch.randn(input_size)  # Batch=1, Channels=1, Height=28, Width=28
    
    print("Input shape:", x.shape)
    print("-" * 50)
    
    # Manual forward pass to see dimensions
    # Conv1: 28x28 -> 24x24 (5x5 kernel reduces by 4)
    x = model.conv1(x)
    print(f"After conv1: {x.shape}")
    
    x = F.relu(x)
    print(f"After relu1: {x.shape}")
    
    # Pool1: 24x24 -> 12x12 (2x2 pooling halves dimensions)
    x = model.pool(x)
    print(f"After pool1: {x.shape}")
    
    # Conv2: 12x12 -> 8x8 (5x5 kernel reduces by 4)
    x = model.conv2(x)
    print(f"After conv2: {x.shape}")
    
    x = F.relu(x)
    print(f"After relu2: {x.shape}")
    
    # Pool2: 8x8 -> 4x4
    x = model.pool(x)
    print(f"After pool2: {x.shape}")
    
    # Flatten: 16*4*4 = 256 features
    x = x.view(-1, 16 * 4 * 4)
    print(f"After flatten: {x.shape}")
    
    # FC layers
    x = model.fc1(x)
    print(f"After fc1: {x.shape}")
    
    x = model.fc2(x)
    print(f"After fc2: {x.shape}")
    
    x = model.fc3(x)
    print(f"After fc3 (output): {x.shape}")

trace_dimensions(model)
```

```python
#Evaluation Metrics

class ModelEvaluator:
    def __init__(self, model, device, class_names):
        self.model = model
        self.device = device
        self.class_names = class_names
        self.reset_metrics()
    
    def reset_metrics(self):
        """Reset all metric trackers"""
        self.all_predictions = []
        self.all_labels = []
        self.all_probabilities = []
        self.losses = []
        
    def evaluate(self, data_loader, criterion):
        """Complete evaluation of the model"""
        self.model.eval()
        self.reset_metrics()
        
        correct = 0
        total = 0
        
        with torch.no_grad():
            for images, labels in data_loader:
                images, labels = images.to(self.device), labels.to(self.device)
                
                # Forward pass
                outputs = self.model(images)
                loss = criterion(outputs, labels)
                
                # Get probabilities (softmax)
                probabilities = torch.softmax(outputs, dim=1)
                
                # Get predictions
                _, predicted = torch.max(outputs, 1)
                
                # Store for later analysis
                self.all_predictions.extend(predicted.cpu().numpy())
                self.all_labels.extend(labels.cpu().numpy())
                self.all_probabilities.extend(probabilities.cpu().numpy())
                self.losses.append(loss.item())
                
                # Count correct
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        
        # Calculate metrics
        metrics = self.calculate_all_metrics()
        return metrics
    
    def calculate_all_metrics(self):
        """Calculate comprehensive metrics"""
        metrics = {}
        
        # 1. Overall Accuracy
        metrics['accuracy'] = accuracy_score(self.all_labels, self.all_predictions)
        
        # 2. Average Loss
        metrics['avg_loss'] = np.mean(self.losses)
        
        # 3. Per-Class Metrics
        metrics['per_class'] = self.calculate_per_class_metrics()
        
        # 4. Confusion Matrix
        metrics['confusion_matrix'] = confusion_matrix(self.all_labels, self.all_predictions)
        
        # 5. Top-k Accuracy
        metrics['top_3_accuracy'] = self.calculate_topk_accuracy(k=3)
        
        # 6. Confidence Metrics
        metrics['confidence'] = self.calculate_confidence_metrics()
        
        return metrics
    
    def calculate_per_class_metrics(self):
        """Calculate precision, recall, F1 for each class"""
        from sklearn.metrics import precision_recall_fscore_support
        
        precision, recall, f1, support = precision_recall_fscore_support(
            self.all_labels, self.all_predictions, average=None
        )
        
        per_class = {}
        for i, class_name in enumerate(self.class_names):
            per_class[class_name] = {
                'precision': precision[i],
                'recall': recall[i],
                'f1_score': f1[i],
                'support': support[i]
            }
        
        return per_class
    
    def calculate_topk_accuracy(self, k=3):
        """Calculate top-k accuracy (is correct label in top k predictions?)"""
        correct = 0
        total = len(self.all_labels)
        
        probabilities = np.array(self.all_probabilities)
        labels = np.array(self.all_labels)
        
        # Get top k predictions for each sample
        top_k_preds = np.argsort(probabilities, axis=1)[:, -k:]
        
        for i in range(total):
            if labels[i] in top_k_preds[i]:
                correct += 1
        
        return correct / total
    
    def calculate_confidence_metrics(self):
        """Analyze model confidence"""
        probabilities = np.array(self.all_probabilities)
        predictions = np.array(self.all_predictions)
        labels = np.array(self.all_labels)
        
        # Get confidence (max probability) for each prediction
        confidences = np.max(probabilities, axis=1)
        
        # Separate correct and incorrect predictions
        correct_mask = predictions == labels
        correct_confidences = confidences[correct_mask]
        incorrect_confidences = confidences[~correct_mask]
        
        return {
            'mean_confidence': np.mean(confidences),
            'mean_confidence_correct': np.mean(correct_confidences) if len(correct_confidences) > 0 else 0,
            'mean_confidence_incorrect': np.mean(incorrect_confidences) if len(incorrect_confidences) > 0 else 0,
            'low_confidence_predictions': np.sum(confidences < 0.5) / len(confidences)
        }
```

```python
class MetricsVisualizer:
    def __init__(self, evaluator, class_names):
        self.evaluator = evaluator
        self.class_names = class_names
    
    def plot_confusion_matrix(self, metrics, normalize=True, figsize=(10, 8)):
        """Plot confusion matrix heatmap"""
        cm = metrics['confusion_matrix']
        
        if normalize:
            cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
            fmt = '.2f'
            title = 'Normalized Confusion Matrix'
        else:
            fmt = 'd'
            title = 'Confusion Matrix'
        
        plt.figure(figsize=figsize)
        sns.heatmap(cm, annot=True, fmt=fmt, cmap='Blues',
                   xticklabels=self.class_names,
                   yticklabels=self.class_names)
        plt.title(title)
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.show()
        
        # Print most confused pairs
        self.print_most_confused(cm)
    
    def print_most_confused(self, cm, top_k=5):
        """Find and print most confused class pairs"""
        # Set diagonal to 0 (we don't care about correct predictions)
        cm_copy = cm.copy()
        np.fill_diagonal(cm_copy, 0)
        
        # Find top k confused pairs
        confused_pairs = []
        for i in range(len(cm_copy)):
            for j in range(len(cm_copy)):
                if i != j:
                    confused_pairs.append((cm_copy[i,j], i, j))
        
        confused_pairs.sort(reverse=True)
        
        print("\nMost Confused Class Pairs:")
        print("-" * 50)
        for count, true_idx, pred_idx in confused_pairs[:top_k]:
            true_class = self.class_names[true_idx]
            pred_class = self.class_names[pred_idx]
            print(f"{true_class}  {pred_class}: {count:.0f} times")
    
    def plot_per_class_metrics(self, metrics, figsize=(12, 6)):
        """Plot per-class precision, recall, and F1 scores"""
        per_class = metrics['per_class']
        
        classes = list(per_class.keys())
        precision = [per_class[c]['precision'] for c in classes]
        recall = [per_class[c]['recall'] for c in classes]
        f1_scores = [per_class[c]['f1_score'] for c in classes]
        
        x = np.arange(len(classes))
        width = 0.25
        
        fig, ax = plt.subplots(figsize=figsize)
        ax.bar(x - width, precision, width, label='Precision', alpha=0.8)
        ax.bar(x, recall, width, label='Recall', alpha=0.8)
        ax.bar(x + width, f1_scores, width, label='F1-Score', alpha=0.8)
        
        ax.set_xlabel('Classes')
        ax.set_ylabel('Score')
        ax.set_title('Per-Class Performance Metrics')
        ax.set_xticks(x)
        ax.set_xticklabels(classes, rotation=45, ha='right')
        ax.legend()
        ax.grid(axis='y', alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        # Print worst performing classes
        self.print_worst_classes(per_class)
    
    def print_worst_classes(self, per_class, metric='f1_score', k=3):
        """Identify worst performing classes"""
        class_scores = [(c, per_class[c][metric]) for c in per_class]
        class_scores.sort(key=lambda x: x[1])
        
        print(f"\nWorst Performing Classes (by {metric}):")
        print("-" * 50)
        for class_name, score in class_scores[:k]:
            support = per_class[class_name]['support']
            print(f"{class_name}: {score:.3f} (n={support})")
    
    def plot_confidence_distribution(self, figsize=(12, 5)):
        """Plot distribution of model confidence"""
        probabilities = np.array(self.evaluator.all_probabilities)
        predictions = np.array(self.evaluator.all_predictions)
        labels = np.array(self.evaluator.all_labels)
        
        # Get max probability (confidence) for each prediction
        confidences = np.max(probabilities, axis=1)
        correct_mask = predictions == labels
        
        fig, axes = plt.subplots(1, 2, figsize=figsize)
        
        # Overall confidence distribution
        axes[0].hist(confidences, bins=50, alpha=0.7, edgecolor='black')
        axes[0].axvline(confidences.mean(), color='red', linestyle='--', 
                       label=f'Mean: {confidences.mean():.3f}')
        axes[0].set_xlabel('Confidence')
        axes[0].set_ylabel('Count')
        axes[0].set_title('Overall Confidence Distribution')
        axes[0].legend()
        axes[0].grid(alpha=0.3)
        
        # Confidence for correct vs incorrect
        axes[1].hist(confidences[correct_mask], bins=30, alpha=0.5, 
                    label='Correct', color='green')
        axes[1].hist(confidences[~correct_mask], bins=30, alpha=0.5, 
                    label='Incorrect', color='red')
        axes[1].set_xlabel('Confidence')
        axes[1].set_ylabel('Count')
        axes[1].set_title('Confidence: Correct vs Incorrect Predictions')
        axes[1].legend()
        axes[1].grid(alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def analyze_misclassifications(self, data_loader, num_examples=5):
        """Show examples of misclassified items"""
        self.evaluator.model.eval()
        misclassified = []
        
        with torch.no_grad():
            for images, labels in data_loader:
                images, labels = images.to(self.evaluator.device), labels.to(self.evaluator.device)
                outputs = self.evaluator.model(images)
                _, predicted = torch.max(outputs, 1)
                
                # Find misclassifications
                wrong_mask = predicted != labels
                wrong_indices = wrong_mask.nonzero(as_tuple=True)[0]
                
                for idx in wrong_indices:
                    if len(misclassified) < num_examples:
                        misclassified.append({
                            'image': images[idx].cpu(),
                            'true_label': labels[idx].item(),
                            'predicted_label': predicted[idx].item(),
                            'probabilities': torch.softmax(outputs[idx], dim=0).cpu()
                        })
                
                if len(misclassified) >= num_examples:
                    break
        
        # Visualize misclassifications
        if misclassified:
            fig, axes = plt.subplots(1, min(num_examples, len(misclassified)), 
                                     figsize=(3*num_examples, 4))
            if num_examples == 1:
                axes = [axes]
            
            for i, item in enumerate(misclassified[:num_examples]):
                img = item['image'].squeeze()
                axes[i].imshow(img, cmap='gray')
                axes[i].set_title(f"True: {self.class_names[item['true_label']]}\n"
                                 f"Pred: {self.class_names[item['predicted_label']]}\n"
                                 f"Conf: {item['probabilities'][item['predicted_label']]:.2f}")
                axes[i].axis('off')
            
            plt.suptitle('Misclassified Examples')
            plt.tight_layout()
            plt.show()
```
